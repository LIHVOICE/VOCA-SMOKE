```python
# defining the mannwhitneyu test
from scipy.stats import mannwhitneyu

def mannwhitneyu_test(cg, sg, included_columns, excluded_columns):

    mannwhitneyu_test = pd.DataFrame()
    p_values = []
    feature_columns = []
    sg_mean_std = []
    cg_mean_std = []
    mean_std = []
    total_samples = pd.concat([cg, sg])

    for column in included_columns:
        if column not in excluded_columns:
            _, p = mannwhitneyu(sg[column], cg[column])
            p_values.append(p)
            feature_columns.append(column)
            sg_mean_std.append(str(round(sg[column].describe()['mean'], 3)) + '(' + str(round(sg[column].describe()['std'], 3)) + ')')
            cg_mean_std.append(str(round(cg[column].describe()['mean'], 3)) + '(' + str(round(cg[column].describe()['std'], 3)) + ')')
            mean_std.append(str(round(total_samples[column].describe()['mean'], 3)) + '(' + str(round(total_samples[column].describe()['std'], 3)) + ')')

    mannwhitneyu_test['features'] = feature_columns
    mannwhitneyu_test['sg_mean_std'] = sg_mean_std
    mannwhitneyu_test['cg_mean_std'] = cg_mean_std
    mannwhitneyu_test['mean_std'] = mean_std
    mannwhitneyu_test['p_values'] = p_values

    return mannwhitneyu_test

"""     significant_features = []
    cg_avgs = []
    sg_avgs = []
    differences = []
    for _, row in mannwhitneyu_test.iterrows():
        if row['p_values'] < 0.05:
            sig_feature = row['features']
            significant_features.append(sig_feature)
            cg_avgs.append(cg[sig_feature].mean())
            sg_avgs.append(sg[sig_feature].mean())
            differences.append(cg[sig_feature].mean() - sg[sig_feature].mean())
            print(row['features'] + ": " + str(row['p_values']))

"""
```

```python
# Applying the HOLM correction on the resulting p_values

from statsmodels.stats.multitest import multipletests

dfs = [sigf1, sigf3, sigf5, sigf1_en, sigf3_en, sigf5_en, sigf1_fr, sigf3_fr, sigf5_fr, sigm1, sigm3, sigm5, sigm1_en, sigm3_en, sigm5_en, sigm1_fr, sigm3_fr, sigm5_fr]

# Combine p-values for correction
p_values = []
for df in dfs:
    p_values.extend(df['p_values'].values)

# Flatten the list of p-values
#flat_p_values = [p for sublist in p_values for p in sublist]

# Apply HOLM correction
alpha = 0.05
reject, corrected_p_values, _, _ = multipletests(p_values, alpha=alpha, method='Holm')
```
